```python
""" imports """

import numpy as np
import os
from xml.dom import minidom
import pandas as pd 
```

```python
path = '/content/drive/MyDrive/Project/dataset'
xml_p = '/content/drive/MyDrive/Project/labels'
bad_samples = ['a01-117-05-02.png', 'r06-022-03-05.png', 'p02-109-01-00.png']
paths_list = []
labels = []
vocab, max_len = set(), 0

# iterate thorugh all the folders
for d1 in os.listdir(path):
  cur_dir1 = os.path.join(path, d1)
  for d2 in os.listdir(cur_dir1):
    cur_dir2 = os.path.join(cur_dir1, d2)

    xml_path = os.path.join(xml_p, d2) + '.xml'
    xml = minidom.parse(xml_path)
    words = xml.getElementsByTagName('word')

    for img in os.listdir(cur_dir2):
      if img in bad_samples:
        continue

      # the image is valid - add it to the list
      paths_list.append(os.path.join(cur_dir2, img)) 

      name = img[:-4] # remove extension .png

      # search for the corrisponding label
      for i in range(len(words)):
        if words[i].attributes['id'].value == name:

          #the label was found - add it to the list and update vocab, max_len.
          label = words[i].attributes['text'].value
          labels.append(label)
          vocab.update(list(label))
          max_len = max(max_len, len(label))
          words.pop(i)
          break
```

```python
df = np.array([paths_list, labels]).T
pd.DataFrame(df).to_csv('/content/drive/MyDrive/Project/data.csv', header = ['path', 'label'], index=None)
```
